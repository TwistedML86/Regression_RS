```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#make a function
def least_squares_regression(x,y):
  n = len(x)

  # calculate Sum
  sum_x = np.sum(x)
  sum_y = np.sum(y)
  sum_xy = np.sum(x*y)
  sum_x2 = np.sum(x**2)

  # calculate slope (b1) and intercepy (bo)
  b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - (sum_x) **2)
  bo = (sum_y - b1 * sum_x) / n

  # Predicted y values
  y_hat = bo + b1 * x       # no residuals yet
  return b1, bo, y_hat

```

```python
x = np.array([1,2,3])
y = np.array([10, 12, 25])

b1, bo, y_hat = least_squares_regression(x,y)

print("Slope (b1):", b1)
print("Intercept (bo):", bo)
print("y_hat =", bo, "+", b1, "*x")
```

```python
# Create a new table with 'charges' (column 6) as 'x' and 'age' (column 0) as 'y'
x_y_table = insurance_data.select(6, 0).relabel(6, 'x').relabel(0, 'y')

# Round the values in the table to two decimal places
x_y_table['x'] = np.round(x_y_table['x'], 2)
x_y_table['y'] = np.round(x_y_table['y'], 2)

# Show the first 5 rows of the new table
x_y_table.show(5)
```

```python
# Calculation with charges in thousands of dollars

# Extract 'age' and 'charges' columns from insurance_data
ages = insurance_data.column('age')
charges_k = insurance_data.column('charges')

# Divide charges by 1000 to change units to thousands of dollars
charges_k = charges_k / 1000

# Calculate correlation and standard deviation
r = np.corrcoef(ages, charges_k)[0, 1]
std_ages = np.std(ages)
std_charges = np.std(charges_k)

# Calculate the slope using the formula: r * (std_y / std_x)
slope = r * (std_charges / std_ages)
print(f'Slope with charges in thousands of dollars: {slope:.4f}')
```

```python
# Calculate the means of age and charges (in thousands)
mean_age = np.mean(ages)
mean_charges = np.mean(charges_k) # charges_k is already in thousands from previous cell

# Calculate the intercept using the formula: bo = mean_y - slope * mean_x
intercept_calculated = mean_charges - slope * mean_age

print(f"Slope of the Regression Line: {slope:.4f}")
print(f"Calculated Intercept: {intercept_calculated:.4f}")
```
